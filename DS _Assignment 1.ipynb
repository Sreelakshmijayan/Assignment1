# i. Countplot

```python
sns.countplot(x='species',data=data1)
plt.show()                                           
```

# ii. Boxplot

```python
plt.figure(figsize=(15,10))
plt.subplot(2,2,1)
sns.boxplot(x='species',y='sepal_length',data=data1)
plt.subplot(2,2,2)
sns.boxplot(x='species',y='sepal_width',data=data1)
plt.subplot(2,2,3)
sns.boxplot(x='species',y='petal_length',data=data1)
plt.subplot(2,2,4)
sns.boxplot(x='species',y='petal_width',data=data1)
pass
```

# iii. Stripplot

```python
plt.figure(figsize=(15,10))
plt.subplot(2,2,1)
sns.stripplot(x=data['species'],y=data['sepal_length'],data=data1,hue='species')
plt.subplot(2,2,2)
sns.stripplot(x=data['species'],y=data['sepal_width'],data=data1,hue='species')
plt.subplot(2,2,3)
sns.stripplot(x=data['species'],y=data['petal_length'],data=data1,hue='species')
plt.subplot(2,2,4)
sns.stripplot(x=data['species'],y=data['petal_width'],data=data1,hue='species')
pass
```

# iv. Heatmap

```python
sns.heatmap(data.corr(),annot=True)
plt.xticks(rotation=90)
```
